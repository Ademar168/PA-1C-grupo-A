```python
#Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique: • 0: Morto • 1 a 30: Crítico • 31 a 70: Normal • 71 a 100: Saudável

energia = int(input("Digite a energia do seu personagem: "))

print(f"Você tem {energia} de energia.")

if (energia >= 71) and (energia <= 100):
    print(f"O seu personagem ainda está com {energia} de energia, ele está saudável")

elif (energia >= 31) and (energia <= 70):
    print(f"O seu personagem ainda está com {energia} de energia, ele está Normal")

elif (energia >= 1) and (energia <= 30):
    print(f"O seu personagem ainda está com {energia} de energia, ele está em estado Critico")

elif energia == 0:
    print(f"Você está com {energia} de energia, você está Morto")
else:
    print("Valor de energia inválido. Por favor, digite um número entre 0 e 100.")
```

```python
#Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras: • Pontuação menor que 5: sem bônus • Pontuação de 5 a 7: bônus de 5% do salário • Pontuação de 8 a 9: bônus de 10% do salário • Pontuação igual a 10: bônus de 15% do salário

salario = float(input("Digite o seu salário: "))
pont = int(input("Digite a pontuação do funcionario: "))
print(f'salario R$: {salario}')

if pont < 5:
    print(f"você recebera R$: {salario} ")

if pont >= 5 and pont <= 7:
    calc = salario * 0.05
    total = salario + calc
    print(f"Você ganhou 5% de bonus e recebera R$: {total}")

if pont >= 8 and pont <= 9:
    calc = salario * 0.10
    total = salario + calc
    print(f"Você ganhou 10% de bonus e recebera R$: {total}")

if pont == 10:
    calc = salario * 0.15
    total = salario + calc
    print(f"Você ganhou 15% de bonus e recebera R$: {total}")
```
