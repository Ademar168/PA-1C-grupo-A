```python
# calculadora simples - escrever um programa que calcula as 04 operaçôes matemática

# opção 1 - somar
# opção 2 - subtrair
# opção 3 - multiplicar
# opção 4 - dividir
# opção 5 - sair

opcao = "" # Changed 'opção' to 'opcao' for consistency with input variable

print ("========== MENU ========")
print ("")

print ("opção 1 - somar")
print ("opção 2 - subtrair")
print ("opção 3 - multiplicar")
print ("opção 4 - dividir")
print ("opção 5 - sair")
print ("=====================")
print ("")

try:

   while opcao != 5:
    opcao = int (input("digite a opção desejada ou 5 para sair:"))

    if opcao == 1:
     a = float (input("digite o primeiro número"))
     b = float (input("digite o segundo número"))

     print (f"total da soma de {a} + {b} é igual: {a + b}")
     print ("")

    elif opcao == 2:
     a = float (input("digite o número"))
     b = float (input("digite o segundo número"))

     print (f"total da subtração de {a} - {b} é igual: {a - b}")
     print ("")

    elif opcao == 3:
     a = float (input("digite o número"))
     b = float (input("digite o segundo número "))

     print(f"total da multiplicação de {a} * {b} é igual: {a * b}")
     print ("")

    elif opcao == 4:
     a = float (input("digite o número"))
     b = float (input("digite o segundo número"))

     # Fixed the f-string syntax for division
     print(f"total da divisão de {a} / {b} é igual: {a / b}")
     print("")

    elif opcao == 5:
     print("Saindo da calculadora. Até mais!")

    else:
     print("Opção inválida. Por favor, escolha uma opção de 1 a 5.")

except ValueError:
  print("Entrada inválida. Por favor, digite um número.")
except ZeroDivisionError:
  print("Não é possível dividir por zero.")
except Exception as e:
  print(f"Ocorreu um erro inesperado: {e}")
```
