fazer um programa que converte a temperatura de celsius para fahrenheit.

Entrada : graus celsius

processamento:converte a temperatura para fahrenheit

saida : mostrar a temperatura em fahrenheit

F=(C x 9/5) +32

```python
# Entrada
print("------------ENTRADA DA TEMPERATURA------------")
celsius = float(input('digite a temperatura em graus celsius'))

# processamento

fahrenheit = (celsius * 9/5) + 32

#saida

print('-------------CONVERÇÃO-------------')
print(f'A temperatura em fahrenheit é: {fahrenheit}')
```

```python
# Escrever um programa que o aluno digita 04 notas dos bimestre e o programa calcula a média e mostra na saida se o aluno foi aprovado, recuperação ou reprovado

nota1 = float (input("digite a nota do 1º bimestre"))
nota2 = float (input("digite a nota do 2º bimestre"))
nota3 = float (input("digite a nota do 3º bimestre"))
nota4 = float (input("digite a nota do 4º bimestre"))

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >=7:

  print(f" você foi aprovado, sua média foi : {media}")

elif media >=5:

  print (f" você está de recuperação, sua média foi : {media}")

else:

  print (f"você foi reprovado, sua media foi : {media}")
```
