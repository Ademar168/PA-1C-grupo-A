fazer um programa que converte a temperatura de celsius para fahrenheit.

Entrada : graus celsius

processamento:converte a temperatura para fahrenheit

saida : mostrar a temperatura em fahrenheit

F=(C x 9/5) +32

```python
# Entrada
print("------------ENTRADA DA TEMPERATURA------------")
celsius = float(input('digite a temperatura em graus celsius'))

# processamento

fahrenheit = (celsius * 9/5) + 32

#saida

print('-------------CONVERÇÃO-------------')
print(f'A temperatura em fahrenheit é: {fahrenheit}')
```
